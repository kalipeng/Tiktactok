Problem Statement for Tic Tac Toe:
Given an existing board of tic tac toe, find the winner of the game.

Sample input into your function:

board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]
Sample output from your function: 'X won', or 'O won', or 'Draw'.





```python

board = [['O', 'X', 'X'],
         ['X', 'X', 'O'],
         ['O', 'X', 'X']]

def win():
    for x in range(3):
        if board[x][0] == board[x][1] == board[x][2]:  # board [0][0]first row, first element
            return board[x][0] # either X or O

    for y in range(3):
        if board[0][y] == board[1][y] == board[2][y]:
            return board[0][y]

    for x in range(0, 3, 2):
        if board[x][0] == board[1][1] == board[2-x][2]:
            return board[x][0]
    return False

if win() is not False:
    print(win()+" won")
```
